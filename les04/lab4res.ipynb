1. **Что такое медиана? Рассчитать медиану для предложенного ряда чисел.**

Медиана — это статистическая мера, которая указывает на центральное значение в упорядоченном ряду чисел. Она делит ряд пополам: половина чисел меньше медианы, а другая половина — больше.

+ Ряд чисел:  67, 89, 45, 56
+ Упорядоченный ряд чисел: 45, 56, 67, 89
+ Т.к. ряд чисел четный, то медианой будет среднеарифметическое двух центральных чисел: (56 + 67) / 2 = 61.5
+ Медиана: 61.5

2. **Что такое дисперсия? Рассчитать дисперсию для предложенного ряда чисел.**

Дисперсия — это статистическая мера, которая показывает, насколько значения в наборе данных отклоняются от среднего значения. Чем больше дисперсия, тем сильнее значения разбросаны относительно среднего.
+ Ряд чисел: 67, 89, 45, 56
+ Среднее значение: (67+89+45+56)/4 = 64.25
+ Значение (xi​)       Отклонение (xi−μ)   Квадрат отклонения ((xi​−μ)^2)
  + ***67***  | 67 - 64.25 = ***2.75*** | 2.752=7.56252.75^2 = 7.56252.752=***7.5625***
  + ***89***  |  89 - 64.25 = ***24.75***| 24.752=612.562524.75^2 = 612.562524.752=***612.5625***
  + ***45***  |  45 - 64.25 = ***-19.25*** | (−19.25)2=370.5625(-19.25)^2 = 370.5625(−19.25)2=***370.5625***
  + ***56***  |  56 - 64.25 = ***-8.25*** |(−8.25)2=68.0625(-8.25)^2 = 68.0625(−8.25)2=***68.0625***
+ Сумма квадратов отклонений: 1058.75
+ Дисперсия: 1058.75 / 4 = 264.6875

4. **Что такое Series и чем он отличается от DataFrame**

*Series* — это одномерная структура данных в библиотеке pandas, похожая на столбец таблицы или одномерный массив. У каждого элемента в Series есть индекс (по умолчанию — числовой, но может быть и строковым, датой и т. д.).

*DataFrame* — это двумерная структура данных, похожая на таблицу в Excel или SQL. Она состоит из строк и столбцов, где каждый столбец — это отдельный Series.

Отличия:
+ Размерность:
  + Series - Одномерная (столбец)
  + DataFrame - Двумерная (таблица)
+ Индексация:
  + Series - Есть индекс для каждого элемента
  + DataFrame - Есть индекс для строк и имена столбцов
+ Структура:
  + Series - Один столбец
  + DataFrame - Несколько столбцов (каждый — Series)
+ Данные:
  + Series - Один тип данных для всех элементов
  + DataFrame - Разные типы в разных столбцах
+ Применение:
  + Series - Для работы с одномерными данными
  + DataFrame - Для работы с табличными данными

6. **Можно ли выполнить операцию сложения над двумя объектами Series, если они имеют:**
+ одинаковые типы данных, одинаковые размеры и одинаковые индексы?
  + Сложение выполнится. Результат — новая Series с теми же индексами, где каждый элемент — сумма соответствующих элементов исходных Series.
+ одинаковые типы данных, одинаковые размеры, но разные индексы?,
  + Сложение выполнится только для совпадающих индексов. Результат — Series с объединёнными индексами, где суммируются только элементы с одинаковыми индексами. Остальные — NaN.
+ одинаковые типы данных, разные размеры и одинаковые индексы?
  + Ошибка. Если размеры разные, но индексы совпадают, pandas не сможет выполнить операцию (несоответствие размеров).
+ разные типы данных, одинаковые размеры и одинаковые индексы?
  + Сложение выполнится, если типы совместимы. Если типы данных можно автоматически привести к общему (например, int + float), результат будет с общим типом. Если типы несовместимы (например, int + str), — ошибка
+ разные типы данных, одинаковые размеры и разные индексы?
  + Аналогично п. 4, но только для совпадающих индексов. Суммируются только элементы с совпадающими индексами, если их типы совместимы. Остальные — NaN.
+ разные типы данных, разные размеры и разные индексы?
  + Ошибка или результат с NaN. Если размеры и индексы не совпадают, а типы данных несовместимы, результат будет содержать NaN для всех несовпадающих пар.

8. **Чем отличается метод iloc от метода loс**
В pandas методы iloc и loc используются для индексации и выбора данных в DataFrame или Series, но они работают по-разному:
+ iloc (integer location-based indexing)
  + Выбор по позициям (индексам строк/столбцов).
  + Использует целочисленные индексы (номера строк и столбцов), начиная с 0.
  + Не зависит от меток строк или столбцов (игнорирует их).
  + Подходит, когда нужно выбрать данные по их позиции в таблице.
+ loc (label-based indexing)
  + Выбор по меткам (индексам строк/столбцов).
  + Использует метки строк и названия столбцов (как они определены в DataFrame).
  + Подходит, когда нужно выбрать данные по именам строк или столбцов.

9. **Можно ли добавить в DateFrame дополнительный Series?**
Да, в DataFrame можно добавить дополнительный Series как новый столбец. Это делается несколькими способами:
+ Присвоение по имени столбца. Просто присвойте Series новому столбцу по его имени. Длина Series должна совпадать с количеством строк в DataFrame.
```
import pandas as pd
# Создаём DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
# Создаём Series
new_series = pd.Series([7, 8, 9])
# Добавляем Series как новый столбец 'C'
df['C'] = new_series
```
+ Использование pd.concat(). Если нужно добавить Series как столбец с сохранением индексов, можно использовать pd.concat().
```
df = pd.concat([df, new_series.rename('C')], axis=1)
```
